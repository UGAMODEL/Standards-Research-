> I want to randomly sample 200 rows from two different excel docs and combine them and randomly shuffle them. 

 

```python
import pandas as pd
import random

# Read the two excel files
df1 = pd.read_excel('10102023combined_and_shuffled_requirements.xlsx')
df2 = pd.read_excel('10102023combined_and_shuffled_standards.xlsx')


print(df2)
```

```python


# Combine the samples
combined = pd.concat([df1, df2])

# Randomly shuffle the combined dataframe
shuffled = combined.sample(frac=1, random_state=42)

# Select only the two desired columns
shuffled_selected = shuffled[['text', 'label','document']] # Replace 'column1' and 'column2' with actual column names

# Save the shuffled dataframe as a CSV file
shuffled_selected.to_csv('embed_data', index=False)

shuffled_selected
```
